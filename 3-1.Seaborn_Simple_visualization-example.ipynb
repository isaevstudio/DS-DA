## Advanced / statistical plotting with Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale = 2, palette = 'viridis')
sns.countplot(data = titanic, x = 'sex', hue = 'pclass') # hue: splitting into subcategories and it takes categorical
plt.show()
```

## Categorical Plots

```python
plt.figure(figsize = (12,8))
sns.set(font_scale = 1.5)
sns.stripplot(data = titanic, x='sex', y='age', jitter=True, hue = 'pclass', dodge=True)#one of the data categorical scatter plot
plt.show()

# hue-categorical subcategory; jitter-overlapping points; dodge-separate the subplots
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale = 1.5)
sns.swarmplot(data=titanic, x='sex', y='age', hue='pclass', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data=titanic, x='sex', y='age', hue='pclass', dodge=True)
sns.swarmplot(data=titanic, x='sex', y='age', hue='pclass', dodge=True, color='black')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data=titanic, x='pclass', y='age', hue='sex', dodge=True, split=True)
plt.show()
# split - combines subcategories into one violin
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.barplot(data=titanic, x='pclass', y='age', hue='sex', dodge=True)
plt.show()

# bar plot calculates the mean age for six groups
# the hight of the bars are the mean, 1st bar the mean age is over 40
# lines - is the confidence interval; 95% of all cases the mean age of 1st class is from this point to that(line)
# ci of 95% and n_boot - 1000 simulations 
# 1 class -segnificant; 2 class - hard to say that the means are segnificant cause they are overlapping
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.pointplot(data=titanic, x='pclass', y='age', hue='sex', dodge=True)
plt.show()
# 1st and 3rd - we can conclude with confidance that the male passengers are older than female passengers
```

## Jointplots / Regression

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
sns.set(font_scale = 1.5)
sns.jointplot(data=titanic, x='age', y='fare', height=8, kind='scatter')
plt.show()

#y axis right frequency is fare distribution
#x axis top frequency is age distribution

# kind='hex', majority passengers are 17-25 y.o. with fare price around 50 
# kde kernel density estimator 
```

```python
# regression plot can tell whether the age and the fare got a relation (do the older passengers 
# paid more than younger passengers)

sns.set(font_scale=1.5)
sns.jointplot(data=titanic, x='age', y='fare', kind='reg', height=8)
plt.show()

# the reg. line is upward sloping so it has a positive relationship. The older pessanger, the higher ticket price
# shaded area around the reg line - 95% of confidance interval for reg line 
```

```python
sns.set(font_scale = 1.5)
sns.lmplot(data=titanic, x='age', y='fare', aspect=1, height=8, hue='sex')
plt.show()

# to create the subplots change the hue to col
```

```python
sns.set(font_scale = 1.5)
sns.lmplot(data=titanic, x='age', y='fare', aspect=1, height=8, col='sex') # subplots by col
plt.show()
```

```python
sns.set(font_scale = 1.5)
sns.lmplot(data=titanic, x='age', y='fare', aspect=1, height=8, row='sex') # subplots by rows
plt.show()
```

```python
# logistic reg
sns.set(font_scale = 1.5)
sns.lmplot(data=titanic, x='age', y='survived', aspect=1, height=8, logistic=True, col = 'sex')
plt.show()

# the graph tells us that the older passenger, the lower chance to survive and vice versa.
# after adding col, we find out that for male passenger the younger the chance to survive is higher
# and for female the older the higher chance to survive
# for male it's import to be a child to survive
```

```python
sns.set(font_scale=1.5)
sns.lmplot(data=titanic, x='age', y='survived', aspect=1, height=8, col='pclass', logistic=True)
plt.show()
```

### 

```python
sns.set(font_scale=1.5)
sns.lmplot(data=titanic, x='age', y='survived', aspect=1, height=8, col='pclass', logistic=False)
plt.show()
```

# MatrixPlots / Heatmaps

Visualize the crosstab data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
pd.crosstab(titanic.sex, titanic.pclass).T
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass), annot=True, fmt='d', cmap='Reds', vmax=150)
plt.show()

# we have an outlier in pclass 3 male (all are around 90 - 150); 347 seems to be strange
# so we add vmax to 150
# it will show very dark anything greater than 150; it gets better to differentiate other 5 groups
# the color in heatmap determine by the size of the groups 
```

```python
# by servival rate

survival_rate = pd.crosstab(titanic.sex, titanic.pclass, values=titanic.survived, aggfunc='mean')
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.heatmap(survival_rate, annot=True, fmt='.2f', cmap='Reds')
plt.show()
```

```python
titanic.corr()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.heatmap(titanic.corr(), cmap='Reds', annot=True)
plt.show()
```
